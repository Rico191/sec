```python
import numpy as np
import matplotlib.pyplot as plt


x = 120
pc = 4
pv = 10 
amostra = 1000
demanda = np.random.normal(100, 10, amostra)
lucro = []

for d in demanda:
    lucro.append(min(x, d) * pv - (pc*x))

# Convert to NumPy array
lucro = np.array(lucro)

# Print profit vector
# print("Profit vector:")
# print(lucro)

import seaborn as sns

sns.histplot(lucro, kde=False, stat="density", linewidth=1)
sns.violinplot(lucro, linewidth=1)


```

```python
def fitness(decision):
    x = decision
    pc = 8
    pv = 100
    amostra = 10000
    demanda = np.random.normal(100, 10, amostra)
    lucro = []
    for d in demanda:
        lucro.append(min(x, d) * pv - (pc * x))
    lucro = np.array(lucro)
    return lucro.mean()
fitness(80)

def simulation():
    in_x = [] 
    out_y = []
    for i in range(50, 120, 1):
        in_x.append(i)
        out_y.append(fitness(i))
    return in_x, out_y

x,y = simulation()
plt.plot(x,y)

```
