```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_investment = 10000  # Initial investment amount
annual_return = 0.08        # Annual expected return
annual_volatility = 0.15    # Annual volatility (standard deviation of returns)
years = 10                  # Number of years
num_simulations = 1000      # Number of simulations

# Generate random returns using normal distribution
returns = np.random.normal(annual_return, annual_volatility, (years, num_simulations))

# Initialize arrays to store portfolio values for each simulation
portfolio_values = np.zeros((years+1, num_simulations))
portfolio_values[0] = initial_investment

# Simulate portfolio value over time
for i in range(1, years+1):
    portfolio_values[i] = portfolio_values[i-1] * (1 + returns[i-1])

# Plot the results
plt.figure(figsize=(10, 6))
for i in range(num_simulations):
    plt.plot(range(years+1), portfolio_values[:, i], lw=0.8, alpha=0.5)

plt.title('Monte Carlo Simulation of Investment Portfolio')
plt.xlabel('Years')
plt.ylabel('Portfolio Value ($)')
plt.grid(True)
plt.show()
```
